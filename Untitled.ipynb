```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("smartcart_customers (1).csv")
```

```python
df.head()
```

```python
df["Income"] = df["Income"].fillna(df["Income"].median())
df["Income"]
```

```python
df["Age"] = 2026-df["Year_Birth"]
```

```python
df["Dt_Customer"] = pd.to_datetime(df["Dt_Customer"],dayfirst = True)
refernce = df["Dt_Customer"].max()
df["Customer_tenure_days"] = (refernce-df["Dt_Customer"]).dt.days
```

```python
df.columns
```

```python
df["Total_Spendings"] = df['MntWines']+df['MntFruits']+df['MntMeatProducts']+df['MntFishProducts']+df['MntSweetProducts']+df['MntGoldProds']
```

```python
df["Total_Children"] = df["Kidhome"]+df["Teenhome"]
```

```python
df["Education"].value_counts()
```

```python
df["Education"] = df["Education"].replace({
    "Basic":"Undergraduate","2n Cycle":"Undergraduate","Master":"Postgraduate","PhD":"Postgraduate","Graduate":"Graduation"
})
```

```python
df["Education"].value_counts()
```

```python
df["Marital_Status"].value_counts()
```

```python
df["Living_with"] = df["Marital_Status"].replace({"Married":"Partner","Together":'Partner','Single':'Alone','Divorced':'Alone','Widow':'Alone','Absurd':'Alone','YOLO':'Alone'})
```

```python
df["Living_with"].value_counts()
```

```python
df.columns
```

```python
cols  = ["ID","Year_Birth","Marital_Status","Kidhome","Teenhome","Dt_Customer",'MntWines', 'MntFruits',
       'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts',
       'MntGoldProds']
df_cleaned = df.drop(columns = cols)
```

```python
df_cleaned.shape
```

```python
# cols = ['Income','Recency','Response','Age','Total_Spendings','Total_Children']
# sns.pairplot(df_cleaned[cols])
```

```python
print(len(df_cleaned))
df_cleaned = df_cleaned[(df_cleaned["Age"]<90)]
df_cleaned = df_cleaned[(df_cleaned["Income"]<600000)]
print(len(df_cleaned))
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder()
cat_cols = ["Education","Living_with"]
enc_cols = ohe.fit_transform(df_cleaned[cat_cols])
```

```python
enc_df = pd.DataFrame(enc_cols.toarray(),columns = ohe.get_feature_names_out(cat_cols),index = df_cleaned.index)
```

```python
encoded_df = pd.concat([df_cleaned.drop(columns = cat_cols),enc_df],axis = 1)
```

```python
encoded_df.shape
```

```python
from sklearn.preprocessing import StandardScaler
X = encoded_df
```

```python
s = StandardScaler()
X_scaled = s.fit_transform(X)
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components = 3)
X_pca = pca.fit_transform(X_scaled)
```

```python
pca.explained_variance_ratio_
```

```python
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111,projection="3d")
plt.scatter(X_pca[:,0],X_pca[:,1],X_pca[:,2])
```

```python
from sklearn.cluster import KMeans
from kneed import KneeLocator
wcss = []
for k in range(1,11):
    kmeans = KMeans(n_clusters = k,random_state = 42)
    kmeans.fit_predict(X_pca)
    wcss.append(kmeans.inertia_)
```

```python
knee = KneeLocator(range(1,11),wcss,curve="convex",direction = "decreasing")
optimal_k = knee.elbow
```

```python
optimal_k
```

```python
plt.plot(range(1,11),wcss,marker='o')
plt.show()
```

```python
from sklearn.metrics import silhouette_score
scores = []
for k in range(2,11):
    kmeans = KMeans(n_clusters = k ,random_state = 42)
    labels = kmeans.fit_predict(X_pca)
    score = silhouette_score(X_pca,labels)
    scores.append(score)
plt.plot(range(2,11),scores,marker='o')
plt.show()
```

```python
kmeans = KMeans(n_clusters = 4,random_state=42)
labels_kmeans = kmeans.fit_predict(X_pca)
```

```python
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111,projection="3d")
plt.scatter(X_pca[:, 0], X_pca[:, 1], X_pca[:, 2], c=labels_kmeans)
```

```python
from sklearn.cluster import AgglomerativeClustering
agg_clf = AgglomerativeClustering(n_clusters = 4,linkage = "ward")
labels_agg = agg_clf.fit_predict(X_pca)
```

```python
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111,projection="3d")
plt.scatter(X_pca[:, 0], X_pca[:, 1], X_pca[:, 2], c=labels_agg)
```

```python
X["clusters"] = labels_agg
# X= X.drop(columns = "labels")
```
